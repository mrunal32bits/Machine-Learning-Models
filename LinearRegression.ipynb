# Linear Regression Model for House Prize Prediction from Area

```python
import pandas as pd                              # imported numpy and pandas to fetch and handle data from  csv file
import numpy as np                               # we put data columns in numpy array before doing operation on it
from sklearn.linear_model import LinearRegression   
import matplotlib.pyplot as plt                    
```

```python
data=pd.read_csv("house.csv")
df=pd.DataFrame(data)
print(df)
```

```python
X=np.array(df["area"])
Y=np.array(df["prize"])
```

```python
plt.style.use("seaborn")
plt.scatter(X,Y,c="Red",s=100,edgecolor="Black")
plt.ylabel("House Prizes")
plt.xlabel("Area in SqFt")
plt.show
```

```python
xm=round(np.mean(X))
ym=round(np.mean(Y))
print("X mean is",xm)
print("Y mean is",ym)
```

```python
X=np.array(df["area"])
Y=np.array(df["prize"])
m=0                                         #m is slpoe of line calculated with help of mean values              
for i in range(0,len(X)):
    m=round(((X[i]-xm)*(Y[i]-ym))/((X[i]-xm)**2))
print("Slope of regression lineis",m)
    
    
```

```python
c=round(ym-xm*m)                 #we found y-intercept with hepl of mean values
print("y-intercept is",c)
```

```python
Yp=[]                              #we find a line which pass through this y coordinates called Y predicted
for i in range(0,len(Y)):            #but this is not regreesion line 
    Yp.append(round(m*X[i]+c))
    
    
print(Yp)    
```

```python
#finding error between Y predicted and Actual Y value 
plt.scatter(X,Yp,s=100,c="Green")
plt.scatter(X,Y,s=100,c="Red")
plt.show()
```

## there are different meathods to minimize the error like root mean square, R squared
## WE ARE USING R SQUARED METHOD

```python
R_sq=0                                          #R**2=sum(Yp-ym)**2/sum(Y-ym)**2
for i in range(0,len(Y)):
    R_sq=((Yp[i]-ym)**2)/((Y[i]-ym)**2)
    
print(R_sq)    
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
reg=linear_model.LinearRegression()
reg.fit(df[["area"]],df.prize)

    
```
